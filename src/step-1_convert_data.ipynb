```python
import pickle
import numpy as np
import mne
data_path = '/home/xiaowjia/data/Friends_fNIRS/data/step1_PPCS_0.01_0.2PictureRating.pkl'
with open(data_path, 'rb') as f:
    data = pickle.load(f)
epoch_data = {}
for key in data.keys():
    hand = int(key[5])
    
    relationship_label = 1 if key[6] in ['1', '2'] else 0
    pair_idx = f"{hand}-{relationship_label}-{key[7:]}"    
    epoch_data.setdefault(hand, {}).setdefault(relationship_label, {}).setdefault(2, {}).setdefault(pair_idx, []).append(data[key]['2'].get_data())
    epoch_data.setdefault(hand, {}).setdefault(relationship_label, {}).setdefault(3, {}).setdefault(pair_idx, []).append(data[key]['3'].get_data())
    
# save data
save_path = '/home/xiaowjia/data/Friends_fNIRS/data/step1_PPCS_0.01_0.2PictureRating_data.pkl'
with open(save_path, 'wb') as f:
    pickle.dump(epoch_data, f)
    
```
